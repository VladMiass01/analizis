##Определите, какой источник трафика наиболее выгоден для компании по текущим данным


```python
df_itog['Просчет стоимости модели'] = pd.to_numeric(df_itog['Просчет стоимости модели'], errors='coerce').fillna(0)
traffic_source_summary = df_itog.groupby('Source')[['Конверсия', 'Просчет стоимости модели']].sum()
traffic_source_summary['Выгода'] = traffic_source_summary['Конверсия'] / traffic_source_summary['Просчет стоимости модели']
most_profitable_source = traffic_source_summary['Выгода'].idxmax()
print(f"Наиболее выгодный источник трафика: {most_profitable_source}")
```

##Ответьте на вопрос: каких показателей не хватает, чтобы посчитать чистую прибыль?

Чтобы посчитать чистую прибыль надо получить данные по затратам.

##Сделать прогноз до конца февраля по количеству конверсий на каждый день

```python
from datetime import datetime, timedelta

conversion_by_date = df_itog.groupby('Date')['Конверсия'].sum()
conversion_by_date.index = pd.to_datetime(conversion_by_date.index)
jan_feb_data = conversion_by_date[(conversion_by_date.index.month == 1) | (conversion_by_date.index.month == 2)]
average_conversions = jan_feb_data.mean()
end_of_february = datetime(2025, 2, 28)
remaining_days = [end_of_february - timedelta(days=i) for i in range(3)]
forecasted_conversions = {day: average_conversions for day in remaining_days}
print("Прогнозируемая сумма конверсий на оставшиеся дни февраля:")
for day, conversions in forecasted_conversions.items():
    print(f"{day.strftime('%Y-%m-%d')}: {conversions:.0f} конверсий")
```

##Какая будет выручка за первый квартал, если средняя стоимость авто останется неизменной, а продажи будут пропорциональны текущим данным?

```python
import numpy as np

sales_data = df_itog[df_itog['Продажа'] == 1]
january_data = sales_data[sales_data['Date'].dt.month == 1]
february_data = sales_data[sales_data['Date'].dt.month == 2]
revenue_january = january_data['цена руб'].sum()
revenue_february = february_data['цена руб'].sum()
revenue_data = pd.DataFrame({
    'Month': [1, 2],
    'Revenue': [revenue_january, revenue_february]
})
coefficients = np.polyfit(revenue_data['Month'], revenue_data['Revenue'], 1)
polynomial = np.poly1d(coefficients)
forecasted_revenue_march = polynomial(3)
print(f"Выручка за январь: {revenue_january:.2f} руб.")
print(f"Выручка за февраль: {revenue_february:.2f} руб.")
print(f"Прогнозируемая выручка за март (линейная экстраполяция): {forecasted_revenue_march:.2f} руб.")
itog_kvartal = revenue_january + revenue_february + forecasted_revenue_march
print(f"Прогнозируемая выручка за квартал: {itog_kvartal:.2f} руб.")
```
